# Exploratory Data Analysis (EDA) on Date Fruit Dataset
This notebook performs exploratory data analysis on the **Date_Fruit_Datasets.xlsx** file.

## 1. Import Required Libraries

```python

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configure visualization styles
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('Set2')
```

## 2. Load the Dataset

```python
!pip install openpyxl

# Load the dataset
df = pd.read_excel('Date_Fruit_Datasets.xlsx')

# Display basic info
df.head()
```

## 3. Dataset Overview

```python

# Display first and last few rows
display(df.head())
display(df.tail())

# Dataset shape
print("Shape of dataset:", df.shape)

# Column names
print("\nColumn Names:")
print(df.columns.tolist())

# Data types
print("\nData Types:")
print(df.dtypes)

# Missing values count
print("\nMissing Values:")
print(df.isnull().sum())

# Unique values count
print("\nUnique Values in each column:")
print(df.nunique())
```

## 4. Data Quality Checks

```python

# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values per column:\n", missing_values)

# Check for duplicates
duplicate_rows = df[df.duplicated()]
print(f"\nNumber of duplicate rows: {duplicate_rows.shape[0]}")

# Identify numeric columns with invalid (negative) values if any
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    invalid = df[df[col] < 0]
    if not invalid.empty:
        print(f"\nInvalid (negative) values found in column: {col}")
```

## 5. Descriptive Statistics

```python

# Summary statistics
df.describe().T
```

## 6. Univariate Analysis

```python

# Plot distribution for numerical columns
df.hist(figsize=(12,10), bins=15, edgecolor='black')
plt.suptitle('Distribution of Numerical Features', fontsize=16)
plt.show()
```

## 7. Correlation Analysis

```python

# Compute correlation matrix
corr = df.corr(numeric_only=True)

# Display heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

## 8. Categorical Feature Analysis

```python

# Identify categorical columns
cat_cols = df.select_dtypes(include=['object']).columns

for col in cat_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

## 9. Outlier Detection

```python

# Boxplots for numeric columns
for col in numeric_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(data=df, x=col)
    plt.title(f'Boxplot of {col}')
    plt.show()
```

## 10. Conclusion


- Dataset overview and structure examined.
- Data quality issues (missing, duplicates, invalids) checked.
- Basic descriptive statistics computed.
- Univariate and correlation analyses visualized.
- Outliers and categorical distributions explored.

This EDA helps understand data distribution, relationships, and potential cleaning steps.
